```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns
from xgboost import XGBRegressor
```

```python
data = pd.read_csv('housing.csv')
```

```python
x = data.drop(columns='MEDV',axis=1)
y = data['MEDV']
```

```python
#print(x)
```

```python
#print(y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1,test_size=0.2)
```

```python
model = XGBRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
train_predict = model.predict(x_train)
print(train_predict)
```

```python
score_1 = metrics.r2_score(y_train,train_predict)
print(score_1)
```

```python
score_error = metrics.mean_absolute_error(y_train,train_predict)
print(score_error)
```

```python
plt.scatter(y_train,train_predict)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.show()
```

```python
test_predict = model.predict(x_test)
print(test_predict)
```

```python
score_r = metrics.r2_score(y_test,test_predict)
print(score_r)
```

```python
score_test_error = metrics.mean_absolute_error(y_test,test_predict)
print(score_test_error)
```

```python
plt.scatter(y_test,test_predict)
plt.xlabel('Actual Prices')
plt.ylabel('Predicted Prices')
plt.show()
```
